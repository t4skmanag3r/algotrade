# Using algotrade library

```python
# importing algotrade library
import algotrade
```

## Creating and testing custom strategy

```python
# Checking available strategies
algotrade.strategies.strategyList()
```

### Defining new strategy

```python
# Strategies and their weights added or subtracted from score
buy_strategies = ["buy_conv", "buy_cloud", "buy_leadspan"]
buy_weights = [1, 4, 3]
sell_strategies = ["sell_conv", "sell_cloud", "sell_leadspan"]
sell_weights = [1, 3, 1]

# Threshold needed to exceed to trigger signals
buy_threshold = 4
sell_threshold = 3

# Initializing new strategy with arguments
strategy = algotrade.testing.NewStrategy(buy_strategies, buy_weights, sell_strategies, sell_weights, buy_threshold, sell_threshold)
```

### Testing initialization

```python
ticker = 'AAPL'
start_date = '2012-01-01'

# Initializing testing with new strategy
test = algotrade.testing.TestStrategy(ticker, start_date, strategy)
```

### Printing Stats

```python
# printing performance statistics of strategy
test.stats
```

### Ploting buy/sell markers

```python
# ploting some latest data with buy/sell dates
days = 1000 # last 1000 days
display_ichimoku = False
display_ma = True

test.plotBuySell(days, display_ichimoku, display_ma)
```
